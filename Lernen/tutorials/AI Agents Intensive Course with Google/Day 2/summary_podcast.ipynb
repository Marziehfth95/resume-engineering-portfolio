#### Die Grundidee
Ein Foundation Model ist im Kern ein sehr starkes Vorhersagesystem für Sprache, aber ohne externe Werkzeuge kann es weder echte Daten abrufen noch Aktionen ausführen. Deshalb ist der große Sprung von „ein Modell antwortet“ zu „ein Agent erledigt Aufgaben“ die Fähigkeit, Tools zu benutzen.

Man kann sich das so vorstellen: Das Modell ist der Kopf, aber Tools sind die Hände, Augen und Ohren. Ein Agent kann dadurch z. B. etwas nachschlagen, eine E-Mail senden, eine Datenbank ändern oder einen Prozess starten



#### Was ist ein Tool
Ein Tool ist einfach eine Funktion oder ein Programm, das das Modell aufruft, wenn es etwas tun soll, was es selbst nicht direkt kann. Dabei gibt es zwei große Zwecke: Das Modell kann entweder etwas Neues erfahren oder etwas in der Welt bewirken.


„Etwas Neues erfahren“ heißt zum Beispiel: Wetter abfragen, Börsenkurs holen, Websuche machen, Datenbank lesen. „Etwas bewirken“ heißt zum Beispiel: E-Mail versenden, Kalendertermin anlegen, Daten aktualisieren oder eine Datei speichern.

#### Drei Tool-Arten
Es werden meist drei Arten von Tools unterschieden. 

Erstens gibt es Function Tools: Das sind vom Entwickler selbst definierte Funktionen, oft mit klaren Dokumentationen und Parametern, damit das Modell weiß, wie es sie richtig aufruft.


Zweitens gibt es Built-in Tools: Die stellt der Anbieter der Modellplattform direkt bereit, etwa Websuche, Codeausführung oder das Abrufen von Inhalten. Diese Tools sind für den Entwickler oft gar nicht als eigene Funktion sichtbar, sondern einfach vorhanden.


Drittens gibt es Agent Tools: Dabei ruft ein Agent einen anderen spezialisierten Agenten auf, so wie ein Manager einen Spezialisten um einen Bericht bittet. Der Hauptagent bleibt dabei in Kontrolle und nutzt das Ergebnis dann für seine eigene Entscheidung.



#### Warum gute Tool Definitionen wichtig sind
Das Modell versteht ein Tool nur über das, was wir ihm über die Beschreibung, den Namen und die Parameter verrätst. Deshalb ist Dokumentation nicht Beiwerk, sondern der eigentliche Vertrag zwischen Modell und Tool.


Ein gutes Tool sollte die Aufgabe beschreiben, nicht den technischen Ablauf. Also lieber „Besprechungsraum buchen“ statt „API-Endpunkt X mit 15 Flags aufrufen“. So bleibt das Tool verständlich und das Modell kann besser entscheiden, wann es passt.


Ein zweiter wichtiger Punkt ist: Tools sollten klar abgegrenzte Aufgaben haben. Wenn ein Tool zu allgemein oder zu komplex ist, wird es für das Modell schwerer, korrekt zu wählen und richtig zu benutzen.

#### Warum kleine Ausgaben besser sind
Ein Tool sollte möglichst knapp zurückgeben, was wirklich nötig ist. Wenn es riesige Datenmengen, lange Logs oder komplette Tabellen in den Modellkontext schiebt, wird der Kontext unnötig voll, teuer und schwerer zu verarbeiten.

Darum ist es oft besser, eine kurze Zusammenfassung, einen Status oder einen Verweis auf eine externe Ressource zurückzugeben. Das ist auch der Grund, warum MCP und ähnliche Systeme mit Referenzen und strukturierten Ergebnissen arbeiten.

#### Fehler richtig behandeln
Fehler sind normal, aber sie müssen für das Modell verständlich sein. Ein Fehler wie „500 Internal Server Error“ hilft einem Agenten kaum weiter, während eine Nachricht wie „Rate limit erreicht, bitte 15 Sekunden warten“ direkt handlungsfähig ist.

* Wichtig: Im MCP Kontext sollen Fehler aus der Tool Ausführung nicht einfach als allgemeiner Protokollfehler verschwinden, sondern als Ergebnis mit isError = true zurückkommen. So sieht das Modell, dass die Aktion fehlgeschlagen ist, und kann selbst eine Korrektur versuchen.



#### Was MCP eigentlich ist
MCP steht für Model Context Protocol. Es ist ein offener Standard, der es Agenten und Tools erleichtert, miteinander zu sprechen, ohne dass für jede Kombination ein eigener Spezial Adapter gebaut werden muss.

Die Idee dahinter ist, die alte „n mal m“-Problematik zu lösen: Früher musste jede Modell-Tool Kombination oft einzeln integriert werden. MCP schafft dafür eine gemeinsame Sprache und macht die Integration modularer und wiederverwendbar.

#### Die drei MCP-Rollen
MCP hat drei Hauptrollen: Host, Client und Server. Der Host ist die Hauptanwendung, in der das Agentenverhalten orchestriert wird, und er ist auch für Guardrails und Sicherheitsregeln verantwortlich.

Der Client ist die Kommunikationsschicht im Host. Er hält die Verbindung, verwaltet die Sitzung und sendet die Tool-Aufrufe an den Server.

Der Server ist der Dienst, der die Tools tatsächlich anbietet. Er beschreibt, welche Tools verfügbar sind, nimmt Anfragen entgegen, führt sie aus und sendet das Ergebnis zurück.

#### Wie MCP kommuniziert
Die Sprache zwischen den Komponenten ist JSON-RPC 2.0. Das bedeutet: Die Nachrichten haben ein standardisiertes Format, damit verschiedene Systeme zuverlässig miteinander sprechen können.

Als Transport nutzt MCP je nach Fall unterschiedliche Wege. Für lokale Prozesse ist oft direkter Prozess-zu-Prozess-Austausch praktisch, für verteilte Systeme ist streamable HTTP wichtig, weil es Remote-Kommunikation und Streaming erlaubt.



#### Tool Definitionen in MCP
Ein MCP Tool hat typischerweise einen Namen, eine Beschreibung und ein Input Schema. Dieses Schema sagt genau, welche Eingaben das Tool erwartet, und optional kann auch ein Output Schema definiert sein.

Das ist wichtig, weil der Host dann nicht raten muss, wie ein Tool benutzt wird. Er kann stattdessen die Struktur prüfen, die Eingaben validieren und die Antwort in einem maschinenlesbaren Format auswerten.



#### Warum MCP nützlich ist
Der größte Vorteil ist, dass Tools und Agenten austauschbarer werden. Man kann neue Tools leichter hinzufügen, Systeme modular aufbauen und verschiedene Anbieter besser kombinieren.

Langfristig kann das zu einem Ökosystem führen, in dem Agenten neue Fähigkeiten dynamisch entdecken und nutzen, ohne dass jede Funktion einzeln hart verdrahtet werden muss. Das macht Agentensysteme flexibler und skalierbarer.

#### Das große Skalierungsproblem
Das Problem ist: Wenn ein Agent theoretisch Hunderte oder Tausende von Tools kennt, kann man nicht einfach alle Tool Beschreibungen in den Prompt laden. Das würde den Kontext sprengen und das Modell eher verwirren als helfen.

* Wichtig: Die Lösungsidee ist Tool Retrieval, also ungefähr wie RAG für Tools: Erst sucht das System semantisch nach den relevantesten Tools und lädt dann nur diese wenigen Kandidaten in den Kontext. So bleibt das Modell fokussiert.

#### Sicherheitsrisiken
MCP löst nicht automatisch alle Sicherheitsprobleme. Besonders wichtig ist das confused deputy Problem: Ein Modell oder Server mit hohen Rechten könnte dazu gebracht werden, eine Aktion auszuführen, die der eigentliche Nutzer gar nicht autorisiert hat.

Das kann durch Prompt Injection oder unklare Berechtigungsgrenzen passieren. Deshalb gilt: Der Standard selbst ist nicht die vollständige Sicherheitslösung, sondern braucht zusätzliche Authentifizierung, Autorisierung und Governance außen herum.

#### Was Unternehmen tun sollten
Für den Enterprise Einsatz sollte MCP nicht direkt und ungeschützt ans Netz gehen. Stattdessen braucht man eine zusätzliche Sicherheits- und Governance Schicht, etwa über API-Gateway, Authentifizierung, Berechtigungsprüfung, Logging und Rate Limiting.

Die wichtige Faustregel lautet: Sicherheit um MCP herum, nicht nur in MCP selbst. Das Protokoll macht die Verbindung möglich, aber die Schutzschicht sorgt dafür, dass die Verbindung sicher bleibt.

#### Ein einfaches Bild dafür
Stell dir vor, das Modell ist ein sehr kluger Assistent im Büro. Ohne Tools kann er nur reden und planen, aber nicht selbst telefonieren, Termine eintragen oder Dokumente abrufen.

MCP ist dann wie ein sauber genormter Büroanschluss, über den dieser Assistent auf viele Geräte zugreifen kann, ohne für jedes Gerät eine eigene Sonderverkabelung zu brauchen. Genau das macht Agenten praktischer und produktionsreif.
